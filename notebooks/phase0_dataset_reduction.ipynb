```python
import json
import os
import random
DATASET_PATH = "D:/Downloads/VISO/coco/car"
IMG_PATH = os.path.join(DATASET_PATH, "train2017")
ANN_PATH = os.path.join(DATASET_PATH, "C:/Users/KimiaMalekizadeh/Downloads/VISO/coco/car/Annotations/instances_train2017.json")
```

```python
with open(ANN_PATH, 'r') as f:
    coco = json.load(f)

print(len(coco['images']), len(coco['annotations']))
```

```python
NUM_IMAGES = 50
selected_images = random.sample(coco['images'], NUM_IMAGES)

selected_image_ids = {img['id'] for img in selected_images}
```

```python
selected_annotations = [
    ann for ann in coco['annotations']
    if ann['image_id'] in selected_image_ids
]
```

```python
small_coco = {
    "images": selected_images,
    "annotations": selected_annotations,
    "categories": coco["categories"]
}
```

```python
OUT_ANN = os.path.join(DATASET_PATH, "C:/Users/KimiaMalekizadeh/Downloads/VISO/coco/car/Annotations/instances_train2017_small.json")

with open(OUT_ANN, 'w') as f:
    json.dump(small_coco, f)

print("Small dataset saved!")
```
