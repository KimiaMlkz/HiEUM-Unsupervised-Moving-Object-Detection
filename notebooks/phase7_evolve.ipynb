```python
def phase7_self_evolution(label_sequence, window=3):
    evolved = []

    for t in range(len(label_sequence)):
        s = max(0, t - window)
        e = min(len(label_sequence), t + window + 1)

        avg = np.mean(label_sequence[s:e], axis=0)
        evolved.append((avg >= 0.5).astype(np.uint8))

    return evolved
```
